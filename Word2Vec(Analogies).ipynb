**WORD ANALOGIES** - In word analogies task we complete the sentence (if a is to b then c is to __). An example is if a man is related to women the a king is to ?. in this we are finding d such tha all are related to each other by the equation **b-a=d-c**

```
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
```

```
!wget "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"

word_vectors= KeyedVectors.load_word2vec_format("/content/GoogleNews-vectors-negative300.bin.gz",binary=True)
```

```
print(word_vectors.vocab)   # this dictionary conatins all the string
```

```
print(word_vectors['man'])
```

```
def predict_words(a,b,c,word_vectors):

  #accepts a triad of words such that a is related to b and c to d
  a,b,c= a.lower(),b.lower(),c.lower()

  #similarity of |b-a|=|d-c|
  max_similarity = -100  # we can assign very negitive value
  d = None
  word = word_vectors.vocab.keys()    # word would contain all the words present in the dictionary(around 450 billion words)
  wa,wb,wc= word_vectors[a],word_vectors[b],word_vectors[c]  # we are extrax=ctig all the values

  #to find d cosine distance between |b-a| and |d-c| should be max
  # for every word in the vocal we would check the value which is giving the maximun value

  for w in word:
    if w in [a,b,c]:
      continue
    
    wv = word_vectors[w]
    sim =cosine_similarity([wb-wa],[wv-wc])

    if sim > max_similarity:
      max_similarity = sim
      d = w
    
  return d
```

```
triad_2=("man","woman","king")
predict_words(*triad_2,word_vectors)
```
